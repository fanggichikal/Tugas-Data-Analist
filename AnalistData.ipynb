```python
ordersdataset_df['order_approved_at'] = pd.to_datetime(ordersdataset_df['order_approved_at'], errors='coerce')

# Drop rows where 'order_approved_at' is NaT (Not a Time)
ordersdataset_df = ordersdataset_df.dropna(subset=['order_approved_at'])

# Grouping by customer_id and calculating RFM metrics
rfm_df = maindata4_df.groupby(by="customer_id", as_index=False).agg({
    "order_approved_at": "max",  # last order date
    "order_id": "nunique",       # number of unique orders
    "price_x": "sum"             # total revenue
})

rfm_df.columns = ["customer_id", "max_order_timestamp", "frequency", "monetary"]

# Calculate recency (days since last order)
rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date
recent_date = ordersdataset_df["order_approved_at"].dt.date.max()  # Get the latest order date
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)

# Drop max_order_timestamp as it's no longer needed
rfm_df.drop("max_order_timestamp", axis=1, inplace=True)

# Show the final dataframe
rfm_df.head()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))

colors = ["#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4"]

# Recency plot
sn.barplot(
    y="recency", 
    x="customer_id", 
    data=rfm_df.sort_values(by="recency", ascending=True).head(5), 
    hue="customer_id",  
    palette=colors, 
    ax=ax[0],
    legend=False  
)
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency (days)", loc="center", fontsize=18)
ax[0].tick_params(axis='x', labelsize=15)

# Frequency plot
sn.barplot(
    y="frequency", 
    x="customer_id", 
    data=rfm_df.sort_values(by="frequency", ascending=False).head(5), 
    hue="customer_id", 
    palette=colors, 
    ax=ax[1],
    legend=False  
)
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)

# Monetary plot
sn.barplot(
    y="monetary", 
    x="customer_id", 
    data=rfm_df.sort_values(by="monetary", ascending=False).head(5), 
    hue="customer_id",
    palette=colors, 
    ax=ax[2],
    legend=False 
)
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15)

plt.suptitle("Best Customer Based on RFM Parameters (customer_id)", fontsize=20)
plt.show()
```
